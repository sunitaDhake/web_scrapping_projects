```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup 
```

```python
url2="https://www.yelp.com/biz/panera-bread-daly-city#reviews"
headers1={'User-Agent':"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5)AppleWebKit/605.1.15 (KHTML, like Gecko)Version/12.1.1 Safari/605.1.15"}
source1=requests.get(url2,headers=headers1).text
```

```python
source1
soup1=BeautifulSoup(source1,"html.parser")
```

```python
print(soup1.prettify())
```

```python
reviews=soup1.find("yelp-react-root").find_all("p")
reviews=[p.text for p in reviews]   
print(reviews)
```
