```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.yelp.com/search?cflt=restaurants&find_loc=San+Francisco%2C+CA"
headers={'User-Agent':"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5)AppleWebKit/605.1.15 (KHTML, like Gecko)Version/12.1.1 Safari/605.1.15"}
source=requests.get(url,headers=headers).text
```

```python
soup=BeautifulSoup(source,"lxml")
```

```python
print(soup.prettify())
```

```python
import re
```

```python
content=soup.find_all("div",class_="padding-t3__09f24__TMrIW padding-r3__09f24__eaF7p padding-b3__09f24__S8R2d padding-l3__09f24__IOjKY border-color--default__09f24__NPAKY")
contentx=soup.find_all("p",class_="css-dzq7l1")
serial_number=[]
name_restaurant=[]
Famous_for=[]
famous_for1=[]
def detail_extraction(content):

            text=[i.text for i in content]
            text2=[i.text for i in contentx]
            for i in text:

                serial_number.append(i.split(".")[0])
                name_restaurant.append(re.search(r"[A-Za-z\s]+",i.split(".")[1])[0].strip())
            for i in text2:
                Famous_for.append(i)
            for i in Famous_for:
              output = re.sub(r"([a-z])([A-Z])", r"\1 \2", i)
              output="".join(output)
              famous_for1.append(output)


            return serial_number ,name_restaurant,famous_for1[2:-1]
```

```python
detail_extraction(content)
```
